```python
def print_general_stats(df_passed, models_passed, problems_passed):
  print('NbP &     solver_name & model & T/O & N/S & Nbt &  avg. t & total')

  nb_problems = len(problems_passed)

  best_result_per_problem = {}
  solvers_passed = df_passed['Solver'].drop_duplicates().tolist()

  for solver in solvers_passed:
    for model in models_passed:
      best_result_per_problem[(solver,model)] = 0

  problem_solutions_all = df_passed.loc[df_passed['Name'].isin(problems_passed)]

  for problem in problems_passed:
    problem_solutions = df_passed.loc[(df_passed['Name'] == problem) & (df_passed[' NbSwaps'] != -1)]
    # the logic goes as follows - if the timeout is not reached it means we found the best solution
    min_time = problem_solutions['TotalTime'].min()
    min_swaps = problem_solutions[' NbSwaps'].min()
    for solver in solvers_passed:
      for model in models:
        if problem_solutions.loc[(problem_solutions['Solver'] == solver) & (problem_solutions['Model'] == model)].shape[0] == 0:
          None
        elif (min_time  == problem_solutions.loc[(problem_solutions['Solver'] == solver) & (problem_solutions['Model'] == model)]['TotalTime'].iloc[0]) & (min_swaps == problem_solutions.loc[(problem_solutions['Solver'] == solver) & (problem_solutions['Model'] == model)][' NbSwaps'].iloc[0]):
          best_result_per_problem[(solver,model)] += 1

  for solver in solvers_passed:
    for model in models_passed:
      collect_timeouts =  problem_solutions_all.loc[(problem_solutions_all['Solver'] == solver) & (problem_solutions_all['Model'] == model) & (problem_solutions_all['IsTimeout'] == True)].shape[0]
      collect_no_solutions =  problem_solutions_all.loc[(problem_solutions_all['Solver'] == solver) & (problem_solutions_all['Model'] == model) & (problem_solutions_all[' NbSwaps'] == -1)].shape[0]
      total_time_all = problem_solutions_all.loc[(problem_solutions_all['Solver'] == solver) & (problem_solutions_all['Model'] == model)]['TotalTime'].sum()
     #average_time_notos = df_passed.loc[(df_passed['Solver'] == solver) & (df_passed['Model'] == model) & (df_passed['IsTimeout'] == False)]['TotalTime'].sum()
      print(str(nb_problems) + ' & ' + '%15s' % solver + ' & ' + '%5s' % model + ' & ' + '%3s' % str(collect_timeouts) + ' & ' + '%3s' % str(collect_no_solutions) + ' & ' + '%3s' % str(best_result_per_problem[(solver,model)]) + ' & ' + '%7s' % "{:.1f}".format(total_time_all / nb_problems) + ' & ' + '%5s' % "{:.1f}".format(total_time_all / 3600))
  print('\n')
```

```python
def print_general_stats_enumerate(df_passed, models_passed, nb_target_plates_passed, size_target_plates_passed, nb_source_plates_passed):
  for nb_target_plates_ex in nb_target_plates_passed:
    for size_target_plates_ex in size_target_plates_passed:
      for nb_source_plates_ex in nb_source_plates_passed:
        problems_ex = df_passed.loc[(df_passed[' NumTargetPlates'] == nb_target_plates_ex) & (df_passed[' SizeTargetPlates'] == size_target_plates_ex) & (df_passed[' NumSourcePlates'] == nb_source_plates_ex)]['Name'].drop_duplicates().tolist()
        print('nb_target_plates: ' + str(nb_target_plates_ex) + ' size_target_plates: ' + str(size_target_plates_ex) + ' nb_source_plates: ' + str(nb_source_plates_ex) + ' nb_problems: ' + str(len(problems_ex)))
        print_general_stats(df_passed, models_passed, problems_ex)
        print('\n')
```

```python
def best_result_aggr(df_passed, models_passed, problems_passed):
  best_result_per_problem = {}
  solvers_passed = df_passed['Solver'].drop_duplicates().tolist()
  for solver in solvers_passed:
    for model in models_passed:
      best_result_per_problem[(solver,model)] = 0

  for problem in problems_passed:
    problem_solutions = df_passed.loc[(df_passed['Name'] == problem) & (df_passed[' NbSwaps'] != -1)]
    # the logic goes as follows - if the timeout is not reached it means we found the best solution
    min_time = problem_solutions['TotalTime'].min()
    min_swaps = problem_solutions[' NbSwaps'].min()
    for solver in solvers_passed:
      for model in models:
        if problem_solutions.loc[(problem_solutions['Solver'] == solver) & (problem_solutions['Model'] == model)].shape[0] == 0:
          None
        elif (min_time  == problem_solutions.loc[(problem_solutions['Solver'] == solver) & (problem_solutions['Model'] == model)]['TotalTime'].iloc[0]) & (min_swaps == problem_solutions.loc[(problem_solutions['Solver'] == solver) & (problem_solutions['Model'] == model)][' NbSwaps'].iloc[0]):
          best_result_per_problem[(solver,model)] += 1
  return best_result_per_problem
```
